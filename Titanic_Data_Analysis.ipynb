**Creating Family Type**

1 - Alone

2 - 4 Small

'>5' - Large

```python
def trans(num):
    if num==1:
        return 'alone'
    elif num>1 and num<5:
        return 'small'
    else:
        return 'large'
```

```python
df['Family Size'].apply(trans)
```

```python
df['Family Type']=df['Family Size'].apply(trans)
```
